```python
import alignment
import landmarks
from deepface import DeepFace
import matplotlib.pyplot as plt
```

```python

filename = 'thomas.jpg'
# get cropped face and landmarks
faces,lmks = landmarks.get_landmarks(filename,'scrfd')
```

```python
aligned_similarity = alignment.get_face_align(faces[0],lmks[0],method='similarity')
aligned_affine = alignment.get_face_align(faces[0],lmks[0],method='affine')
aligned_rotate = alignment.get_face_align(faces[0],lmks[0],method='rotate')
```

```python
plt.figure()

f,ax = plt.subplots(1,3)

ax[0].imshow(aligned_similarity)
ax[0].axis('off')
ax[0].set_title('similarity')
ax[1].imshow(aligned_affine)
ax[1].axis('off')
ax[1].set_title('affine')
ax[2].imshow(aligned_rotate)
ax[2].axis('off')
ax[2].set_title('rotate')
plt.show()
```
